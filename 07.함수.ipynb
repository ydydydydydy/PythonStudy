```python
def cal(a,b,c):
    if c == '+':
        result = a+b
#         return a+b
    elif c =='-':
        result = a-b
    else:
        result = '연산할 수 없습니다'
        
    return result
    
    
```

# 독스트링(docstring)

 - 함수의 기능 설명 작성
 - shift + tab 누르면 설명 확인 가능
 - 작성하는 방법 : 삼중 문자열로 작성 """ """

```python
def addNum1(num1, num2):
    """ 설명문 작성하는 란 , int 매개변수 2개 받아 더하는 함수입니다."""
    return num1+num2
```

```python
addNum1(10,11)
```

```python
# 파이썬은 동적 자료형으로 자료형 명시를 안해서 타입과 맞지 않는 타입이 들어와도 가능

# 문자열 + 숫자형 --> X 
# 매개변수로 문자열, 숫자형을 받는 것은 가능
#잘못된 매개변수의 자료형을 넣어 함수를 사용 못하는 경우가 빈번
```

```python
#타입 힌트를 작성 
# 힌트이다 보니 자료형 강제 하는 것은 아님
# 타입 힌트 주는법 :   매개변수 : 타입 
def addNum2(num1 : int, num2 : int):
    return num1+num2
```

```python
addNum2('s','ㄴ')
```

```python
#반환되는 타입 힌트 -> 강제성이 있는 것은 아님
#  함수명(매개변수)->타입 :
def addNum3(num1:int, num2 : int) -> int:
    return num1+num2
```

```python
addNum3('s','s')
```

# 매개변수의 기본값 설정

```python
def addNum4(num1 , num2 =11):
    return num1+num2
```

```python
#매개변수의 기본값이 설정 되어있는 함수에서
# 기본값이 있는 매개변수를 작성하지 않을 시
# 설정한 기본값으로 연산
addNum4(10)
```

```python
#num2의 숫자를 바꿔주고 싶다면?
#일반 메소드처럼 사용!
addNum4(10,30)
```

```python
#가변 매개변수
#매개변수의 개수가 변할 수 있는 메소드
#함수 호출 당시에 몇개의 인수가 전달 될 지 모를때

def addNum5(*args):
    for i in args:
        print(i)
```
