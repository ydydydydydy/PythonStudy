# 치환 - 값을 바꾸는 작업

```python
num1 = 10
num2 = 7

# num1을 7로 바꿔주고 num2를 10으로 바꿔주도록 하겠습니다
#값을 임시로 들고있을 변수 --temp
temp = 0

print("바뀌기 전")
print(num1)
print(num2)

temp = num1 #temp = 10
num1 = num2 #num1 = 7 
num2 = temp
print("바뀐 후")
print(num1)
print(num2)
```

```python
num1, num2 = num2,num1
print(num1)
print(num2)
```

```python
#비교연산자 --결과값은 bool 표시 (True, False)
#종류 부등호(< ,<=) 
#같다 의 의미인 ==
#같지 않다 의 의미인 !=

num1 = 3
num2 = 7

print(num1 > num2) #False
print(num1 <= num2) #True
print(num1 == num2) #False
print(num1 != num2) #True
```

```python
#논리 연산자 
# 논리값을 뒤집는 역할 -not 
print(not num1<num2) #False
print(not num1 == num2) #True
```

```python
#and 두 값이 모두 True일 경우만 True
print(num1<num2 and 20==20) # True
print(num1>num2 and 20 == 20)#False
```

```python
#or 두 값중 하나라도 True인 경우 True
print(num1<num2 or 20==20) 
print(num1>num2 or 20 == 20)
print(num1>num2 or 20!=20)
```

```python
#멤버 연산자 - in, not in - True,False
#문자열, 리스트, 튜플 에서 값이 있는지 여부를 확인
#사용하는 방법 - 원하는 문자 in 문자열(변수)
str1 = "안녕하세요"
print("안녕" in str1)
print("hello" not in str1)
```
