```python
#1번
countOfMask = int(input("마스크 개수 입력 >>"))
count_8 = countOfMask //8
temp = countOfMask %8
count_5= temp//5 
temp_5 = temp%5

if temp_5 > 0:
    count_5+=1

print(f"8개들이 포장지 개수 : {count_8}")
print(f"5개들이 포장지 개수 : {count_5}")
```

```python
age = int(input("나이를 입력하세요 >>"))
#유아기 - 6세 미만
#아동기 - 6~10세미만
#청소년기 10~ 24세미만
#성인 - 24~ 40세미만
#장년 - 40 ~65세미만
#노인 - 65세 이상
# 6< age <10 (0)
if age < 6:
    print("유아기 입니다.")
elif age <10:
    print("아동기 입니다.")
elif age <24:
    print("청소년 입니다.")
elif age <40:
    print("성인입니다.")
elif age <65:
    print("장년입니다.")
elif age >=65:
    print("노인입니다.")
```

```python
str1 = ''
if age < 6:
    str1 = "유아기"
elif age <10:
    str1 = "아동기"
elif age <24:
    str1 = "청소년기"
elif age <40:
    str1="성인"
elif age <65:
    str1 = "장년"
elif age >=65:
    str1 = "노인"

    
print(f"{str1} 입니다")
```
