```python
# 클래스는 (변수, 함수) 로 이루어져 있으며
# 클래스는 설계도
# 클래스 문법 
# class 클래스명:
#들여쓰기 변수, 함수 
#변수는 소문자 시작
#약속 클래스명은 대문자 시작 

class People:
    name = ''
    age = 0
    
    def walk(self): 
        #클래스 내 에서 함수 생성시 첫번째 매개변수는 무조건 self
        #self == this
        # 나 자신(class)을 뜻하는 키워드
        
        #self를 사용하는 곳
        # 1.인스턴스(객체) 변수 접근 
        # 2.같은 클래스에서 다른 메소드 호출 
        # 3.다른 클래스 변수, 메소드 호출 
        print(f"{self.name} 님이 이동합니다.") # 1번 경우 
    def eat(self):
        self.walk()  # 2번 경우 
        print('먹습니다')
```

```python
#설계도를 바탕으로 객체 만들어보도록 하겠습니다
#객체 생성 방법
#  객체명 = 클래스명()

wonho = People()
```

```python
print(wonho) #people --> name, age, walk(), eat()
#wonho라는 객체 
#wonho라는 객체 안에 name, age, walk, eat
#name가져와보기

# 가지고오는 방법 객체명.변수명, 객체명.함수명
wonho.name = '원호'
print(wonho.name)

wonho.walk()
```

```python
#객체생성 
#이름 = 클래스명()
ryan = People()
ryan.name = '라이언'
print(ryan.name)
ryan.age = 10
print(ryan.age)
ryan.walk()
```

```python
class PiggyBank:
    #필드
    money =0
    
    #함수
    def deposit(self, input_money):
        self.money+= input_money
        return self.money
    def withdraw(self, withdraw_money):
        if self.money <withdraw_money:
            print('잔액이 부족합니다.')
        else :
            self.money -= withdraw_money
        
    def showMoney(self):
        print(self.money)
```

```python
#객체 지향적으로 프로그램 개발
piggy = PiggyBank()
# print(piggy.money)
# piggy.deposit()

while True:
    choice = int(input("1.저금 2.인출 3.금액확인 4.종료"))
    if choice == 4:
        print('종료합니다')
        break
    elif choice == 1:
        input_money = int(input("저금할 금액을 입력하세요"))
        result= piggy.deposit(input_money)
        print(result)
    elif choice == 2:
        withdraw_money = int(input("출금할 금액을 입력하세요 >>"))
        piggy.withdraw(withdraw_money)
        
    elif choice == 3:
        piggy.showMoney()
    else :
        print("잘못 눌렀습니다.")
```
