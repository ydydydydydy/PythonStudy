```python
random_num = random.randrange(1,51)
count = 1
while True:
    input_num = int(input("숫자를 입력하세요"))
   
    if input_num > 50:
        print("정신 차리세요 50보다 작은수입니다.")
    
    elif input_num < random_num:
        print(f'{input_num}보다 큰 수입니다.')
    elif input_num > random_num:
        print(f'{input_num}보다 작은 수입니다.')
    elif input_num == random_num:
        print(count)
        print("정답입니다!")
        break
   
    count+=1

```

```python
current_weight = int(input("현재 몸무게 : "))
goal_weight = int(input("목표 몸무게 : "))
count = 1
while True:
    minus = int(input(f"{count}주차 감량 몸무게 : "))
    current_weight-=minus # current_weight = current_weight -minus
    
    if current_weight <goal_weight:
        print(f'{current_weight}kg 달성 축하합니다!')
        break
    count+= 1
```

# for문

```python
#for문 문자열, 리스트, 튜플이 들어갔을때 안에 있는 요소를 하나씩 반복

#for문 문법
#for 임의의 변수 in 문자열(or 리스트 or 튜플):
#들여쓰기 

#문자열
for a in 'hello':
    print("hello") #hello가 총 5번 출력
    
    #ㄴ 그 이유는?????? 
    # 1.문자열에 있는 h가 a변수에 한번들어가서 출력
    # 2. 문자열 e ->a변수에 대입 후 출력
    # 3. l -> a 출력
    #4. l -> a print
    #5. o -> a 
    print(a) # 문자열 한개씩 출력
    
    
```

```python
#for - list사용 예시

food_list = ['탕수육', '햄버거', '피자']
#리스트의 경우에는
#리스트에 있는 요소 하나하나가 임의의 변수에 들어가게됨!
for i in food_list:
    print('hello') #3번 출력
    # ㄴ 그 이유는?
    #1. 탕수육 -> i print
    #2. 햄버거 -> i print
    #3. 피자 -> i print
    print(i)
```

```python
#간단한 예제 -for문 list예시를 응용하여 

score_list = [90,45,70,60,55]
cnt = 0
for i in score_list:
    cnt+=1
    
    # for문 안에 if 작성
    if i >=60:
        print(f'{cnt}번 학생은 {i}점으로 합격 입니다.')
    else:
        print(f'{cnt}번 학생은 {i}점으로 불합격 입니다.')
    
    
    
```

```python
#range()함수사용하여 해당 "범위"만큼
# ㄴ 숫자를 만들어내는 함수

#range(시작할 숫자, 종료할 숫자 미만, 증가량)
#range(1, 10, 1) -> 1부터 9까지 1개씩 증가

#감소도 가능
#range(10, 1, -1) -> 10 ~ 2까지 1씩 감소

for i in range(1,10,1):
    print(i)
    
#range(시작할 숫자, 종료할 숫자 미만) - 증가량 1씩인 경우 생략가능

for i in range(1,10):
    print(i)
    #위 반복문과 같은 결과!
    
#range(종료할 숫자 미만) - 시작할 숫자가 0인 경우
for i in range(10):
    print(i)
    
```

```python
list2 = [4,5,2,1,99,15,2,7,27]
# max_num = 0
#list2에 있는 값이 모두 음수라면??
#list2에 없는 값인 max_num이 가장 max값
#list요소중에 하나를 변수화

max_num = list2[0]

for i in list2:
    if max_num <i:
        max_num = i
print(f'가장 큰 수 는 : {max_num}')
```

```python
list2 = [4,5,2,1,99,15,2,7,27]
max_num = list2[0]

#그렇다면 최초로 비교할 때
# list2[0] <list2[0] --> list2[0] <list2[1]

for i in range(1,len(list2)): # range(1,9)-> 1~8 
    if max_num <list2[i]:
        max_num = list2[i]
print(f'가장 큰 수 는 : {max_num}')
```

```python
list2 = [4,5,2,1,99,15,2,7,27,-10]
min_num = list2[0]

#그렇다면 최초로 비교할 때
# list2[0] <list2[0] --> list2[0] <list2[1]

for i in range(1,len(list2)): # range(1,9)-> 1~8 
    if min_num >list2[i]:
        min_num = list2[i]
print(f'가장 작은 수 는 : {min_num}')
```

# 약수 :  1~ 나자신까지 수 중에 나눠지는 수

#### 6의 약수 : 1,2,3,6
##### 왜???? 1부터 6 까지 다 나눠봤지~



```python
num = int(input("정수 입력 >> "))

print(f'{num}의 약수 :' ,end = ' ')
for i in range(1,num+1):
    if num %i == 0:
        print(f'{i}', end = ' ')
```

# 다중 반복문

#### 다중반복문이란? 
###### 반복문 안에 반복문이 존재하는 구문

```python
food_list = ['치킨','피자','콜라']

for i in food_list:
    print(i)
    # 유의사항은 임의의 변수명은 중복되지 않게
    for j in i:
        print(j)
```

```python
list1 = [[10,20,30], [40,50,60],[70,80,90]]

for i in list1:
    print(i)
    for j in i:
        print(j)
```
