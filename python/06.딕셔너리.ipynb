#### key와 value값으로 이루어져 있습니다

## key값은 변하지 않은 고유 값

###### immutable key, mutable value

###### 리스트와 튜플과 달리 순서가 없는 집합

```python
# 딕셔너리 구조 - K : V
#{key : value}
#해당 key, value의 타입은 딱히 정해져 있지 않습니다.

#리스트 == [], 튜플 == () 딕셔너리 == {}

dict1 = {} # 비어 있는 딕셔너리

#str key : int value
dict2 = {"first" : 1, "second" : 2}

#int key : str value
dict3 = {1 : 'one', 2: 'two',3:'three'}

#str key : str value
dict4 = {'dog' :'개', 'cat':'고양이','pig':'돼지'}


print(dict4)
print(type(dict4))
```

```python
# value값이 복합적인 dict

my_info_dict = {'name' : '원호', 'age' :20, 'hobby': ['운동','베이킹']}

print(my_info_dict)
```

```python
# 딕셔너리 값 추가
# 문법
# 딕셔너리명[key값] = value
my_info_dict['addr'] = '금호동'
print(my_info_dict)
```

```python
#딕셔너리 값 가져오기
#문법
# 딕셔너리명[key]
print(my_info_dict['name'])
```

```python
#딕셔너리 값 가져오기2
#문법
# 딕셔너리명.get(key값)
print(my_info_dict.get('hobby'))

```

```python
# 2가지 딕셔너리 가지고 오는 방법의 차이점

# 없는 키값 작성해보독 하겠습니다

# [key] 방식
print(my_info_dict['birth']) # key error
```

```python
#get 방식
print(my_info_dict.get('birth'))


## 결론 : get을 추천한다
# 그 이유 : 에러를 내는 코드가아니라서
```

```python
# key값만 모두 가져오기
#문법
#딕셔너리명.keys()

print(my_info_dict.keys())
print(type(my_info_dict.keys()))

#반복문으로 하나씩 가져오기 가능
for i in my_info_dict.keys():
    print(i)
```

```python
#value 값만 가져오기
# 문법
#딕셔너리명.values()

print(my_info_dict.values())

#반복문으로 하나씩 가져오기 가능
for i in my_info_dict.values():
    print(i)
```

```python
#둘다 가져오기
#문법
#딕셔너리명.items()

print(my_info_dict.items())

#반복문으로 출력하기
for i in my_info_dict.items():
    print(i)
    for j in i:
        print(j)
```

```python
#
for i,j in my_info_dict.items():
    print(i, j )
```

```python
#딕셔너리에서 in 키워드 사용
#딕셔너리의 경우 key값만 가능 
#   ㄴ 해당 key값이 dict안에 있습니까?

# 결과값은 True/False

#문법
#key값 in 딕셔너리명

print('name' in my_info_dict)
print('birth' in my_info_dict)
```

```python
# 딕셔너리 값 삭제
# 문법
# del 딕셔너리명[key]

del my_info_dict['addr']

print(my_info_dict)
```

```python
#딕셔너리 모든 값 삭제
#문법
#딕셔너리명.clear()

my_info_dict.clear()

print(my_info_dict)
```
