```python
#리스트 내부에 있는 값 가지고 오기!

list_int = [3,10,2,50,7]
#문자열의 인덱싱 -- 0번부터 시작합니다.
#리스트의 인덱싱 -- 0번부터 시작합니다.
print(list_int[0]) #3
print(list_int[2]) #2
print(list_int[0] +list_int[2]) # 5
print(type(list_int[0]))
print(type(list_int))
```

```python
#리스트의 값 추가!
#1. appned() 
# 사용법 : 리스트명.append(값) 
# 결과 : 맨 끝에 값이 추가

list3 = [4,7,8,10]
print(f"추가 되기 전 : {list3}")
list3.append(13)
print(f"추가 된 후 : {list3}")
```

```python
#추가하는 방법 22
# 2. insert()
#사용법 : 리스트명.insert(인덱스 위치, 값)
#결과 :작성한 인덱스 위치에 값이 추가 된 후 뒤에 인덱스들은 뒤로 밀립니다!
#  ㄴ 원하는 위치에 값 추가
list4 = ['hello','my','name','is','wh']
print(f"추가 되기 전 : {list4}")
list4.insert(1,'world')
print(f"추가 된 후 : {list4}")
```

```python
#추가하는 방법 33
# 3. extend()
# 사용법 : 리스트명.extend(리스트)
# 결과 : 끝에 여러개(리스트에 있는 값들)을 한번에 추가
list5 = ['사과','딸기','바나나']
list6 = ['망고','수박','메론','키위']
print(f"추가 되기 전 : {list5}")
list5.extend(list6)
print(f"추가 된 후 : {list5}")
```

```python
#값 수정
#수정하는 방법 : 인덱스로 접근 후 값 대입
list7 = [1,3,5,6,9]
print(f"수정 되기 전 : {list7}")
list7[3] =7
print(f"수정 된 후 : {list7}")
```

```python
#값 삭제
#1. del 키워드 사용 
#사용법 : del 리스트[인덱스]
list8 = [0,1,2,3,4,5]
print(f"삭제 되기 전 : {list8}")
del list8[0]
print(f"삭제 된 후 : {list8}")
```

```python
#2. remove() 사용
#사용법 : 리스트명.remove(값)
list9 = ['a','b','c','d','e']
print(f"삭제 되기 전 : {list9}")
list9.remove('d')
print(f"삭제 된 후 : {list9}")

#응용🤣 
if 'd' in list9:
    list9.remove('d')
else:
    print("삭제 할 수 없습니다")
```

```python
#리스트의 길이를 알려주는 함수
#len() -length의 약자
#사용법 : len(리스트명)
#결과는 리스트의 값 개수를 반환
list10 = [1,2,3,4,5]
i = len(list10)
print(i)
print(type(i))
```

```python
#sort() -정렬
#사용법 : 리스트명.sort()
#결과는 리스트에 있는 값을 오름차순으로 정렬

list11 = [9,77,13,51,100,3]
print(f"정렬 되기 전 : {list11}")
list11.sort()
print(f"정렬 된 후 : {list11}")
```

```python
#sort() 문자열도 가능할까?
list_str = ['apple','watermelon','strawberry','banana','kiwi']
print(f"정렬 되기 전 : {list_str}")
list_str.sort()
print(f"정렬 된 후 : {list_str}")
```

```python
#reverse()
# 리스트에 있는 값을 역순으로 뒤집어줌
# 리스트명.reverse()
# ㄴ 결과 : 순서를 뒤집어 줍니다.

list12 = [9,77,13,51,100]
print(f"reverse 되기 전 : {list12}")
list12.reverse()
print(f"reverse 된 후 : {list12}")

# 내림차순으로 정렬
#  sort() ->reverse()
list12.sort()
list12.reverse()
print(f"sort & reverse 된 후 : {list12}")
```

```python
# 리스트의 값 위치 반환 및 제거
# 리스트의 값 위치 반환 -index()
# 리스트명.index(요소)
#결과 : 찾고자 하는 요소의 인덱스 값을 반환

list13 = ['a','b','c','d','e']

index = list13.index('b')
print(index)

print(list13[index])

#반환 후 제거 - pop()
# 마지막 값을 반환 후에 리스트에서 제거
print(list13.pop())
print(list13)
```

```python
#리스트/ 튜플의 슬라이싱 
#문자열 슬라이싱과 동일하게
#[시작할 인덱스 : 끝나는 인덱스 미만]

list1 = [10,20,30,40,50]

print(list1[2:4]) #30, 40

#문자열 슬라이싱과 마찬가지로 0부터 시작시 생략 가능
print(list1[:4]) # == list1[0:4] 

# 끝까지 출력시 마지막인덱스 생략 가능
print(list1[3:])# 40, 50

# 둘 다 생략시 -- 모두
print(list1[:])
```
