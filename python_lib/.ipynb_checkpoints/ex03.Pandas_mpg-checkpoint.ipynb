### 컬럼 -- grade (등급)
- 평균값이 25이상이면 A등급
- 20이상 B등급
- 그 외 C등급

```python
mpg_grade = np.where(mpg_copy['평균'] >= 25, 'A등급', 
                     np.where(mpg_copy['평균']>=20, 'B등급','C등급'))
```

```python
mpg_copy['grade'] = mpg_grade
mpg_copy
```

```python
mpg_copy['grade'].value_counts()
```

## apply()라는 함수
- np.where의 조건식 복잡해지는 것을 간단하게 작성 할 수 있도록 하는 함수

```python
# apply() 사용법

# 매개변수가 함수
# 함수를 만들어서 함수명만 작성 해주시면 됩니다.
def makeGrade(df): # 함수 실행시 조건이 될 시리즈가 필요! 매개변수로 df 작성
    grade = df['평균']
    
    if grade >= 25:
        return "A등급"
    elif grade >= 20:
        return "B등급"
    elif grade >= 15:
        return "C등급"
    else:
        return "F등급"
```

- axis = 0(행) or 1(열, 축)

```python
# apply()
mpg_copy["grade2"] = mpg_copy.apply(makeGrade, axis=1)
mpg_copy
```

```python
mpg_copy['grade2'].value_counts().sort_index()
```

```python
# group by("그룹화할 대상")
grouped_data = mpg_copy.groupby('제조사').mean()
```

```python
# agg() - 다중집계작업
# agg(결과이름 = (컬럼명, 통계함수)) -- 컬럼명과 통계함수는 모두 문자열
# sum, mean, max, min, count
```

```python
# 제조사별 도시연비 평균
mpg_copy.groupby('제조사').agg(city_mean=('도시','mean'))
```

```python
mpg_copy.groupby('제조사').agg(city_max=('도시','max'))
```

```python
mpg_copy.groupby('제조사').agg(city_min=('도시','min'))
```

```python
mpg_copy.groupby('제조사').agg(city_count=('도시','count'))
```

```python
# 여러개 사용도 가능! - agg 사용해서
# 도시평균, 고속도로평균

# groupby로 제조사별로 묶기
# agg사용
mpg_copy.groupby('제조사').agg(city_mean = ('도시','mean'), high_mean = ('고속도로','mean'))
```

```python
# 카테고리별 도시연비
# 카테고리 조사
mpg_copy['category'].value_counts()
```

```python
# 카테고리별로 groupby
# 도시 연비 평균
mpg_copy.groupby('category').agg(city_mean = ('도시','mean'))
```

```python
# 해당 곃과값 정렬
# error! -> 'by' argument
# by - 어떤 컬럼을 바탕으로 
mpg_copy.groupby('category').agg(city_mean = ('도시','mean')).sort_values(by = 'city_mean', ascending=False)
```

```python
# 제조사 별로 그룹을 묶은 후
# 제조사 별 등급(grade컬럼)을 나누기
# 예- audi A등급, B등급이 몇개인가?

# 그룹을 다시 나눌땐
# groupby시 [] 한번 더 감싸서 진행!
mpg_copy.groupby(['제조사','grade']).agg(grade_cnt = ('grade', 'count'))
```

```python
# pd.merge()

# db - join inner, outer - left right
# left = df1이 기준
# right = df2가 기준
# outer - 데이터를 모두 합치는데 만약 df1이나 df2와 일치하지 않는 데이터가 존재시
#        ㄴ nan 값이 들어가게 됨
# test3로 확인

# inner - 교집합 부분만 표시 (일치하지 않는 데이터는 merge X)

# on - 기준이 될 컬럼
# pd.merge(df1, df2 how = '')
```

```python
test1 = pd.DataFrame({'id':[1,2,3,4,5,6],'midterm':[60,80,70,90,85,80]})
test2 = pd.DataFrame({'id':[1,2,3,4,5],'final':[70,83,65,95,80]})
```

```python
test3 = pd.merge(test1, test2, how = 'outer', on = 'id')
test3
```

```python
test4 = pd.merge(test1, test2, how = 'inner', on = 'id')
test4
```

## 2 NaN이란 (not a number)
-  값이 없다

  #### 결측치란?
  - 누락된 값, 비어있는 값
  - 수료과정에서 발생한 오류 때문에 포함될 경우가 종종 발생
  - 결측치 한다는 의미는 aaply() 함수 적용하는데 어려움이 생기고 분석하는데 결과의 오류 발생
  - 결측치 처리하는 방법은 2가지(삭제, 값을 채워주기)

```python
df = pd.DataFrame({'gender':['m','f',np.nan,'m','f'],
                  'score':[5,4,3,4,np.nan]})
df
```

```python
# .dropna() - nan 값인 데이터를 모두 날려버립니다
df.dropna()
```

```python
#.dropna(subset = '기준이 되는 컬럼명')
# score의 값 중 nan값인 데이터만 drop하기
df.dropna(subset='score')
```

```python
# 내가 원하는 행만 삭제
# .drop(내가 삭제할 행(인덱스), axis = 0 ) : axis=0(행),1(열)
df.drop(2, axis = 0)
```

```python
# 여러개 열 삭제
df.drop([2,4], axis = 0)
```

```python
# 내가 원하는 열만 삭제
# .drop(내가 삭제할 컬럼명, axis = 1 )  : axis = 0(행), 1(열)
df.drop('gender', axis = 1)
```

```python
df.drop(['gender','score'], axis = 1)
```
