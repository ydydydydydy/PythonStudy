### Matplotlib 

#### 데이터를 한눈에 볼 수 있게 도표나 차트로 정리를 도와주는 라이브러리
#### 시각화를 통해 데이터의 특징을 파악 가능
#### 분석 결과를 상대방에 효과적으로 전달 가능

#### bar(count), scatter, line

#### pie, histogram, boxplot

#### matplotlib에 존재하는 sub 패키지

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Line Plot 

```python
# line plot을 그리기 위해서는 x축 y축이 필요
x = np.arange(1,6) # x = [1, 2, 3, 4, 5]
np.random.seed(3)
y = np.random.randint(1, 10, size = 5) # y = [?,?,?,?,?]

print(x)
print(y)
```

```python
# plt.plot(x축, y축)

plt.plot(x, y)
```

```python
# plt.plot(y축)
# -- x축은 자동으로 0~ y축의 개수 만큼 증가
plt.plot(y)
```

```python
print(x)
print(y)
```

```python
plt.plot(x, y, ls = ":")
```

```python
plt.plot(x, y, ls = ":", color = 'r', marker = 'D', mfc = 'b')
# plt.show()
# 만약 그래프가 안 나오면 plt.show() 사용해주기
# 쥬피터 노트북은 출력을 지원하는 개발환경이라 사용 안 해줘도 출력
```

```python
# 한 화면에 여러개의 그래프를 출력

# 방법:plt.subplot(row, col, index)

plt.subplot(2, 2, 1)
plt.plot(x, y, ls = '-', c = 'b', marker = "D", ms = 10, mec = 'c', mew = 4, mfc ='y')

plt.subplot(2, 2, 2)
plt.plot(x, y, ls ='--', c = 'm', lw = 5)

plt.subplot(2, 2, 3)
plt.plot(x, y, ls = '-.')

plt.subplot(2, 2, 4)
plt.plot(x, y, ls = '-')
```

```python
# 여러 그래프를 한 창에 그려지게(그래프 겹치기)

one_x = np.arange(1, 4) #[1, 2, 3]
one_y = np.array([1, 2, 3])
one_y2 = np.array([4, 5, 1])
one_y3 = np.array([3, 6, 2])

plt.plot(one_x, one_y)
plt.plot(one_x, one_y2)
plt.plot(one_x, one_y3)
```

```python
# plt.show()
plt.plot(one_x, one_y)
plt.show()

plt.plot(one_x, one_y)
plt.plot(one_x, one_y2)
plt.plot(one_x, one_y3)
```

```python
# title - plt.title('타이틀 적기')
print(x)
print(y)

plt.title('Test title')
plt.plot(x, y)
```

### title()

```python
# title - plt.title('타이틀 적기')
print(x)
print(y)

plt.plot(x,y)
plt.title('Test title')
```

### grid()

```python
# 격자눈금 함수
#plt.grid()

plt.plot(x,y)
plt.title('Test title')
plt.grid()
```

### legend()

```python
test_x = np.arange(1,6)
np.random.seed(5)
test_y = np.random.randint(1,10,size=5)

test_x1 = np.arange(1,6)
np.random.seed(10)
test_y1 = np.random.randint(1,10,size=5)

print(test_x)
print(test_y)
print(test_x1)
print(test_y1)
```

```python
# legend() 사용방법
# plt.legend()
# 단, plt.plot을 그릴 때 해당 plot의 label 값을 지정해주어야 한다
#      ㄴ plt.plot(x, y, label = '라벨값 지정')

plt.plot(test_x, test_y, label = 'y')
plt.plot(test_x1, test_y1, label = 'y1')
plt.legend(loc = 'upper right')
```

### xlabel(), ylabel()

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
# plt.xlabel('이름')
# plt.ylabel('이름')

plt.plot(test_x, test_y, label = 'y')
plt.plot(test_x1, test_y1, label = 'y1')
plt.legend(loc = 'upper right')

plt.xlabel('x축')
plt.ylabel('y축')
```

###  xlim(), ylim()

```python
# xlim, ylim(시작, 끝)

plt.plot(test_x, test_y, label = 'y')
plt.plot(test_x1, test_y1, label = 'y1')
plt.legend(loc = 'upper right')

plt.xlabel('x축')
plt.ylabel('y축')

# x, y이 문자열인 경우 문자로 작성
plt.xlim(0, 10)
plt.ylim(0, 20)
```

### xticks, yticks

```python
# 사용법 xticks, yticks(list, 배열)

plt.plot(test_x, test_y, label = 'y')
plt.plot(test_x, test_y1, label = 'y1')
plt.legend(loc = 'upper right')

plt.xlabel('x축')
plt.ylabel('y축')

plt.xticks(range(1,5))  # 1, 2, 3, 4, 5
plt.yticks(range(1,11,2))  #1, 3, 5, 7, 9
```

## 실습) 장래 인구 변동 데이터

```python
# 컬럼 하나를 인덱스화
fu_pop = pd.read_csv('data/장래_인구변동_KOSIS.csv', encoding='euc-kr', index_col = '인구변동요인별')
fu_pop
```

```python
# .columns - 해당 df 컬럼들을 가지고 올 수 있다
fu_pop.columns
fu_pop_x = list(fu_pop.columns)

print(fu_pop_x)
```

### 연도별 출생아수 

```python
# 데이터 접근
fu_pop_y = fu_pop.iloc[1]
print(fu_pop_y)

fu_pop_y2 = fu_pop.iloc[2]
print(fu_pop_y2)
```

```python
plt.plot(fu_pop_x, fu_pop_y)
```

##
  - 출생아수, 사망자수 그래프 

```python
plt.plot(fu_pop_x, fu_pop_y, marker = '*', ms = 10, label = '출생아수(천명)')
plt.plot(fu_pop_x , fu_pop_y2 , marker='o',label = '사망자수(천명)')
plt.title('국가통계 요인별 장래 인구 동향')
plt.xlabel('연도')
plt.ylabel('사람수')
plt.legend()
plt.grid()

plt.yticks(range(150, 751 , 50))
```

### 그래프 이미지화

```python
plt.plot(fu_pop_x, fu_pop_y, marker="*", ms = 10, label='출생아수')
plt.plot(fu_pop_x, fu_pop_y2, marker=".", ms = 10, label='사망자수')
plt.legend(loc = 'upper right')
plt.title('국가통계 요인별 장래 인구 동향')
plt.legend()
plt.grid()
plt.xlabel('연도')
plt.ylabel('사람수')

plt.yticks(range(150,751, 50))

# plt.savefig('경로/파일명.png', dpi = , bbox_inches)
# dpi = dots per inch : 해상도 설정
# bbox_inches = 저장할 부분의 공간을 지정 = tight 그림의 빈 공간을 최소화

plt.savefig('./test1.png', dpi = 600, bbox_inches = 'tight')
```
