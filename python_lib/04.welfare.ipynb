# marriage 컬럼 데이터 정보
  - 0 : 미해당 (18세 미만)
  - 1 : 결혼
  - 2 : 사별
  - 3 : 미혼
  - 4 : 별거
  - 5 : 미혼
  - 6 : 기타

###  1 - marriage
###  3 - divorce
###  5 - not marriage
###  0, 2, 4, 6 - etc

#####  실습) apply 활용해서 바꾸어 주기

```python
def getMarriage(row):
    marriage = row
    if marriage == 1:
        return "marriage"
    elif marriage == 3:
        return "divorce"
    elif not marriage == 5:
        return "not marriage"
    else:
        return "etc"
```

```python
welfare_copy['marriage'] = welfare_copy['marriage'].apply(getMarriage)
welfare_copy
```

```python
welfare_copy['marriage'].value_counts()
```

```python
sns.countplot(data  = welfare_copy, x = 'marriage')
```

```python
# income == NaN, code_job == NaN 인 값들 drop()
# drop(인덱스, axis = 0)

drop_index = welfare_copy[(welfare_copy['income'].isna()) & (welfare_copy['code_job'].isna())].index
welfare_drop = welfare_copy.drop(drop_index)
welfare_drop.info()
```

```python
welfare_drop
```

```python
# 인덱스 정리
welfare_drop.index = range(1, len(welfare_drop)+1)
welfare_drop
```

```python
# 결측치 채우기
def fill_income(row):
    # 수입 값이 결측치 일때 
    if np.isnan(row['income']):
        return code_median.loc[row['code_job']].values[0]
    # 수입 값이 결측치 아닐 때
    else:
        return row['income']
```

```python
welfare_drop['income'] = welfare_drop.apply(fill_income, axis = 1)
```

```python
welfare_drop.query('income.isna()')
```

```python
welfare_drop = welfare_drop.dropna(subset = ['income'])
welfare_drop.query('income.isna()')
```
