## apply()라는 함수
- np.where의 조건식 복잡해지는 것을 간단하게 작성 할 수 있도록 하는 함수

```python
# apply() 사용법

# 매개변수가 함수
# 함수를 만들어서 함수명만 작성 해주시면 됩니다.
def makeGrade(df): # 함수 실행시 조건이 될 시리즈가 필요! 매개변수로 df 작성
    grade = df['평균']
    
    if grade >= 25:
        return "A등급"
    elif grade >= 20:
        return "B등급"
    elif grade >= 15:
        return "C등급"
    else:
        return "F등급"
```

- axis = 0(행) or 1(열, 축)

```python
# apply()
mpg_copy["grade2"] = mpg_copy.apply(makeGrade, axis=1)
mpg_copy
```

```python
mpg_copy['grade2'].value_counts().sort_index()
```

# NaN이란 (not a number)
 - 값이 없다
 
 #### 결측치란?
  - 누락된 값, 비어있는 값
  - 수집과정에서 발생한 오류 때문에 포함 될 경우가 종종 발생
  - 결측치 한다는 의미는 apply() 함수 적용하는데 어려움이 생기고
  - 분석하는데 결과의 오류 발생
  - 결측치 처리 하는 방법은 2가지 (삭제, 값을 채워주기)  

```python
df = pd.DataFrame({'gender':['m','f',np.nan,'m','f'],
                  'score':[5,4,3,4,np.nan]})
df
```

```python
# .dropna(subset = '기준이 되는 컬럼명')
# score의 값 중 nan값인 데이터만 drop하기
df.dropna(subset='score')
```

```python
# 내가 원하는 행만 삭제
# .drop(내가 삭제할 행(인덱스), axis = 0) : axis = 0(행), 1(열)
df.drop(2, axis = 0)
```

```python
# 여러개 열 삭제
df.drop([2,4], axis = 0)
```

```python
# 내가 원하는 열만 삭제
# .drop(내가 삭제할 컬럼명, axis = 1)   : axis = 0(행), 1(열)
df.drop('gender', axis = 1)
```

```python
df.drop(['gender', 'score'], axis = 1)
```

### @실습 category가 suv인 차들의 제조사를 묶고
###    해당 제조사들의 도시 연비 평균을 구한 후 barplot으로 그려보시오

```python
#1. 카테고리 suv인 차 가지고 오기
mpg_copy[mpg_copy['category'] == 'suv']

# 1.2 query()함수 사용
mpg_suv = mpg_copy.query('category == "suv"')
mpg_suv
```

```python
# 2. 제조사별로 묶기
mpg_suv.groupby('제조사').mean(True)
```

```python
# 3. 제조사들의 도시 연비 구하기
mpg_suv_city = mpg_suv.groupby('제조사', as_index = False).agg(city_mean=('도시', 'mean'))
```

```python
# 4. barplot 그려보기
sns.barplot(data = mpg_suv_city, x = '제조사', y = 'city_mean')
```
